# DBSCAN (Density-Based Spatial Clustering of Applications with Noise)

## Теория

DBSCAN — это алгоритм **кластеризации на основе плотности**. В отличие от K-Means, он не требует заранее задавать число кластеров и способен находить кластеры произвольной формы, а также автоматически определять шумовые точки (выбросы).

### Ключевые понятия

Алгоритм вводит три типа точек:

- **Ядровые точки** (core points) — точки, в $\varepsilon$-окрестности которых содержится не менее `min_samples` точек (включая саму точку).
- **Граничные точки** (border points) — точки, не являющиеся ядровыми, но попадающие в $\varepsilon$-окрестность хотя бы одной ядровой точки.
- **Шумовые точки / выбросы** (noise points) — точки, не являющиеся ни ядровыми, ни граничными; не принадлежат ни одному кластеру.

### Алгоритм

1. Для каждой непосещённой точки найти все точки в её $\varepsilon$-окрестности.
2. Если таких точек не менее `min_samples` — точка является ядровой, начать формирование нового кластера.
3. Рекурсивно добавить в кластер все точки, достижимые по плотности от ядровой точки (то есть все ядровые точки в окрестности, их соседей и т.д.).
4. Если точка не попала ни в один кластер — пометить как шум.

### Ключевые гиперпараметры

| Параметр | Значение по умолчанию | Описание |
|---|---|---|
| **eps** | 0.5 | Радиус $\varepsilon$-окрестности — максимальное расстояние между двумя точками, при котором они считаются соседями |
| **min_samples** | 5 | Минимальное число точек в $\varepsilon$-окрестности для признания точки ядровой |
| **metric** | `'euclidean'` | Метрика расстояния между точками (`'euclidean'`, `'manhattan'`, `'cosine'` и др.) |

## Датасет

В работе используются два датасета.

### make_moons — синтетические данные

Генератор `make_moons` из библиотеки `sklearn` создаёт синтетические 2D-данные в форме двух переплетающихся полумесяцев. Это **невыпуклые** кластеры, которые K-Means не способен корректно разделить. DBSCAN справляется с ними естественно благодаря плотностному подходу.

Параметры генерации:
- `n_samples=300` — 300 точек,
- `noise=0.05` — небольшой шум для реалистичности.

### Iris — реальные данные

Классический датасет **Iris** из библиотеки `sklearn` содержит измерения цветков ириса трёх видов.

В датасете:
- **150 объектов** — образцы цветков (по 50 на каждый вид),
- **4 признака** — морфологические измерения цветка (в см),
- **3 класса** — виды ириса (используются только для оценки качества кластеризации).

### Описание признаков Iris

| Признак | Перевод | Описание |
|---|---|---|
| **sepal length (cm)** | длина чашелистика | длина наружного листочка околоцветника |
| **sepal width (cm)** | ширина чашелистика | ширина наружного листочка околоцветника |
| **petal length (cm)** | длина лепестка | длина внутреннего листочка околоцветника |
| **petal width (cm)** | ширина лепестка | ширина внутреннего листочка околоцветника |

### Описание классов Iris

| Класс | Название | Особенности |
|---|---|---|
| **0 — setosa** | ирис щетинистый | самый мелкий вид; лепестки значительно короче и уже |
| **1 — versicolor** | ирис разноцветный | промежуточные размеры; частично перекрывается с virginica |
| **2 — virginica** | ирис виргинский | самый крупный вид; наибольшие длина и ширина лепестков |


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.datasets import make_moons, load_iris
from sklearn.cluster import DBSCAN, KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import NearestNeighbors
from sklearn.metrics import adjusted_rand_score
```

```python
# Генерация синтетических данных make_moons
X_moons, y_moons = make_moons(
    n_samples=300,
    noise=0.05,
    random_state=42
)

print("Размер данных make_moons:", X_moons.shape)
print("Количество точек в каждом классе:", np.bincount(y_moons))

# Визуализация исходных данных
plt.figure(figsize=(8, 5))
plt.scatter(X_moons[:, 0], X_moons[:, 1], c=y_moons, cmap='tab10', edgecolor='black', s=50, alpha=0.8)
plt.xlabel('Признак 1')
plt.ylabel('Признак 2')
plt.title('Синтетические данные make_moons (два полумесяца)')
plt.tight_layout()
plt.show()
```

```python
# Загрузка датасета Iris
iris = load_iris()

X_iris = iris.data
y_iris = iris.target

feature_names = iris.feature_names
target_names = iris.target_names

print("Размер матрицы признаков X:", X_iris.shape)
print("Размер целевой переменной y:", y_iris.shape)
print("\nНазвания признаков:", feature_names)
print("\nНазвания классов:", target_names)
```

```python
df = pd.DataFrame(X_iris, columns=feature_names)
df["target"] = y_iris

df.head()
```

### Базовые статистики датасета

```python
df.describe().round(2)
```

```python
class_counts = pd.Series(y_iris).value_counts().sort_index()

print("Количество объектов каждого класса:")
for class_id, count in class_counts.items():
    print(f"  {target_names[class_id]} ({class_id}): {count}")
```

```python
class_means = df.groupby("target")[list(feature_names)].mean().T
class_means.columns = [target_names[int(c)] for c in class_means.columns]
class_means = class_means.round(2)

print("Средние значения признаков по классам:\n")
print(class_means.to_string())
print(f"\nSetosa имеет значительно меньшие лепестки (petal), чем остальные два вида.")
print(f"Versicolor и virginica различаются в основном по размерам лепестков и чашелистиков.")
```

## Сравнение K-Means и DBSCAN на make_moons

Покажем принципиальное различие двух алгоритмов на невыпуклых данных. K-Means делит пространство на выпуклые области вокруг центроидов и не может корректно разделить переплетающиеся полумесяцы. DBSCAN находит кластеры по плотности и справляется с этой задачей естественно.

```python
# K-Means на make_moons
kmeans_moons = KMeans(n_clusters=2, init='k-means++', n_init=10, random_state=42)
kmeans_moons.fit(X_moons)
labels_kmeans = kmeans_moons.labels_

# DBSCAN на make_moons
dbscan_moons = DBSCAN(
    eps=0.2,          # радиус окрестности
    min_samples=5     # минимальное число точек
)
dbscan_moons.fit(X_moons)
labels_dbscan = dbscan_moons.labels_

# Информация о кластерах DBSCAN
n_clusters_dbscan = len(set(labels_dbscan)) - (1 if -1 in labels_dbscan else 0)
n_noise_dbscan = np.sum(labels_dbscan == -1)
print(f"DBSCAN: найдено кластеров = {n_clusters_dbscan}, шумовых точек = {n_noise_dbscan}")
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# K-Means
axes[0].scatter(X_moons[:, 0], X_moons[:, 1], c=labels_kmeans, cmap='tab10', edgecolor='black', s=50, alpha=0.8)
axes[0].scatter(
    kmeans_moons.cluster_centers_[:, 0],
    kmeans_moons.cluster_centers_[:, 1],
    c='red', marker='X', s=200, zorder=5, label='Центроиды'
)
axes[0].set_xlabel('Признак 1')
axes[0].set_ylabel('Признак 2')
axes[0].set_title('K-Means (n_clusters=2) — не справляется')
axes[0].legend()

# DBSCAN — шумовые точки отдельным цветом
unique_labels = sorted(set(labels_dbscan))
colors = plt.cm.tab10(np.linspace(0, 1, max(len(unique_labels), 2)))

for label, color in zip(unique_labels, colors):
    mask = labels_dbscan == label
    if label == -1:
        # Шумовые точки — серые крестики
        axes[1].scatter(X_moons[mask, 0], X_moons[mask, 1],
                        c='grey', marker='x', s=60, alpha=0.7, label='Шум')
    else:
        axes[1].scatter(X_moons[mask, 0], X_moons[mask, 1],
                        color=color, edgecolor='black', s=50, alpha=0.8, label=f'Кластер {label}')

axes[1].set_xlabel('Признак 1')
axes[1].set_ylabel('Признак 2')
axes[1].set_title(f'DBSCAN (eps=0.2, min_samples=5) — {n_clusters_dbscan} кластера, {n_noise_dbscan} шумовых')
axes[1].legend()

plt.suptitle('K-Means vs DBSCAN на make_moons', fontsize=14)
plt.tight_layout()
plt.show()
```

## Влияние параметра eps

Параметр `eps` задаёт радиус окрестности. Его выбор критически важен:
- **Малое eps** — мало точек попадает в окрестность; большинство точек становятся шумом, кластеры дробятся.
- **Большое eps** — почти все точки попадают в окрестность друг друга; весь датасет сливается в один кластер.

Исследуем поведение DBSCAN на make_moons при разных значениях eps (min_samples=5 фиксирован).

```python
eps_values = [0.1, 0.2, 0.3, 0.5, 1.0]
min_samples_fixed = 5

fig, axes = plt.subplots(1, len(eps_values), figsize=(20, 4))

for ax, eps in zip(axes, eps_values):
    model = DBSCAN(eps=eps, min_samples=min_samples_fixed)
    labels = model.fit_predict(X_moons)

    n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
    n_noise = np.sum(labels == -1)

    # Цвета: шум — серый, кластеры — tab10
    unique_labels = sorted(set(labels))
    color_map = plt.cm.tab10(np.linspace(0, 1, max(len([l for l in unique_labels if l >= 0]), 1)))
    cluster_idx = 0

    for label in unique_labels:
        mask = labels == label
        if label == -1:
            ax.scatter(X_moons[mask, 0], X_moons[mask, 1],
                       c='grey', marker='x', s=30, alpha=0.6)
        else:
            ax.scatter(X_moons[mask, 0], X_moons[mask, 1],
                       color=color_map[cluster_idx], edgecolor='black', s=30, alpha=0.8)
            cluster_idx += 1

    ax.set_title(f'eps={eps}\nкластеров: {n_clusters}, шума: {n_noise}')
    ax.set_xlabel('Признак 1')
    ax.set_xticks([])
    ax.set_yticks([])

axes[0].set_ylabel('Признак 2')
plt.suptitle(f'Влияние eps на DBSCAN (min_samples={min_samples_fixed})', fontsize=14)
plt.tight_layout()
plt.show()

print(f"\n{'eps':>6}  {'кластеров':>10}  {'шумовых точек':>15}")
print('-' * 36)
for eps in eps_values:
    model = DBSCAN(eps=eps, min_samples=min_samples_fixed)
    labels = model.fit_predict(X_moons)
    n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
    n_noise = np.sum(labels == -1)
    print(f"{eps:>6.1f}  {n_clusters:>10}  {n_noise:>15}")
```

## Влияние параметра min_samples

Параметр `min_samples` задаёт минимальное число точек в $\varepsilon$-окрестности, необходимое для признания точки ядровой:
- **Малое min_samples** — более «мягкий» критерий ядровости; почти все точки становятся ядровыми, шума почти нет.
- **Большое min_samples** — более «строгий» критерий; только точки в плотных областях считаются ядровыми, остальные уходят в шум.

Исследуем поведение DBSCAN на make_moons при разных значениях min_samples (eps=0.2 фиксирован).

```python
min_samples_values = [3, 5, 10, 15, 20]
eps_fixed = 0.2

fig, axes = plt.subplots(1, len(min_samples_values), figsize=(20, 4))

for ax, min_s in zip(axes, min_samples_values):
    model = DBSCAN(eps=eps_fixed, min_samples=min_s)
    labels = model.fit_predict(X_moons)

    n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
    n_noise = np.sum(labels == -1)

    unique_labels = sorted(set(labels))
    color_map = plt.cm.tab10(np.linspace(0, 1, max(len([l for l in unique_labels if l >= 0]), 1)))
    cluster_idx = 0

    for label in unique_labels:
        mask = labels == label
        if label == -1:
            ax.scatter(X_moons[mask, 0], X_moons[mask, 1],
                       c='grey', marker='x', s=30, alpha=0.6)
        else:
            ax.scatter(X_moons[mask, 0], X_moons[mask, 1],
                       color=color_map[cluster_idx], edgecolor='black', s=30, alpha=0.8)
            cluster_idx += 1

    ax.set_title(f'min_samples={min_s}\nкластеров: {n_clusters}, шума: {n_noise}')
    ax.set_xlabel('Признак 1')
    ax.set_xticks([])
    ax.set_yticks([])

axes[0].set_ylabel('Признак 2')
plt.suptitle(f'Влияние min_samples на DBSCAN (eps={eps_fixed})', fontsize=14)
plt.tight_layout()
plt.show()

print(f"\n{'min_samples':>12}  {'кластеров':>10}  {'шумовых точек':>15}")
print('-' * 42)
for min_s in min_samples_values:
    model = DBSCAN(eps=eps_fixed, min_samples=min_s)
    labels = model.fit_predict(X_moons)
    n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
    n_noise = np.sum(labels == -1)
    print(f"{min_s:>12}  {n_clusters:>10}  {n_noise:>15}")
```

## DBSCAN на датасете Iris

Применим DBSCAN к датасету Iris. Перед кластеризацией выполним масштабирование с помощью `StandardScaler` — DBSCAN использует расстояния, поэтому масштаб признаков важен. Визуализируем результат в проекции на два признака: длина лепестка и ширина лепестка.

```python
# Масштабирование признаков Iris
scaler = StandardScaler()
X_iris_scaled = scaler.fit_transform(X_iris)

print("Среднее до масштабирования (первые 4 признака):", np.round(X_iris.mean(axis=0), 2))
print("Среднее после масштабирования:", np.round(X_iris_scaled.mean(axis=0), 2))
print("Стд. откл. после масштабирования:", np.round(X_iris_scaled.std(axis=0), 2))
```

```python
# Применение DBSCAN к Iris
dbscan_iris = DBSCAN(
    eps=0.5,          # радиус окрестности
    min_samples=5     # минимальное число точек
)

dbscan_iris.fit(X_iris_scaled)
labels_iris = dbscan_iris.labels_

n_clusters_iris = len(set(labels_iris)) - (1 if -1 in labels_iris else 0)
n_noise_iris = np.sum(labels_iris == -1)

print(f"Найдено кластеров: {n_clusters_iris}")
print(f"Шумовых точек:     {n_noise_iris}")
print("\nРаспределение меток:", dict(zip(*np.unique(labels_iris, return_counts=True))))
```

```python
# Визуализация: кластеры DBSCAN vs истинные метки
# Индексы признаков: petal length=2, petal width=3
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Предсказанные кластеры DBSCAN
unique_labels_iris = sorted(set(labels_iris))
color_map_iris = plt.cm.tab10(np.linspace(0, 1, max(len([l for l in unique_labels_iris if l >= 0]), 1)))
cluster_idx = 0

for label in unique_labels_iris:
    mask = labels_iris == label
    if label == -1:
        axes[0].scatter(X_iris[mask, 2], X_iris[mask, 3],
                        c='grey', marker='x', s=80, alpha=0.8, label='Шум', zorder=5)
    else:
        axes[0].scatter(X_iris[mask, 2], X_iris[mask, 3],
                        color=color_map_iris[cluster_idx], edgecolor='black', s=60, alpha=0.8,
                        label=f'Кластер {label}')
        cluster_idx += 1

axes[0].set_xlabel('Длина лепестка (petal length)')
axes[0].set_ylabel('Ширина лепестка (petal width)')
axes[0].set_title(f'DBSCAN: {n_clusters_iris} кластера, {n_noise_iris} шумовых точек')
axes[0].legend()

# Истинные метки классов
scatter = axes[1].scatter(
    X_iris[:, 2], X_iris[:, 3],
    c=y_iris, cmap='tab10', edgecolor='black', s=60, alpha=0.8
)
axes[1].set_xlabel('Длина лепестка (petal length)')
axes[1].set_ylabel('Ширина лепестка (petal width)')
axes[1].set_title('Истинные метки классов')
plt.colorbar(scatter, ax=axes[1], label='Класс')

plt.suptitle('DBSCAN на Iris: кластеры vs истинные классы', fontsize=14)
plt.tight_layout()
plt.show()
```

```python
# Оценка качества кластеризации
ari_iris = adjusted_rand_score(y_iris, labels_iris)
print(f"Adjusted Rand Index (ARI): {ari_iris:.4f}")
print("(ARI=1.0 — идеальное совпадение, ARI=0 — случайное)")

# Таблица соответствия: истинные метки vs кластеры DBSCAN
print("\nТаблица соответствия (истинный класс → кластер DBSCAN):")
comparison = pd.crosstab(
    pd.Series(y_iris, name='Истинный класс').map(dict(enumerate(target_names))),
    pd.Series(labels_iris, name='Кластер DBSCAN')
)
print(comparison)
```

## Подбор параметра eps (метод k-расстояний)

Один из практических способов выбрать `eps` — **метод k-расстояний** (k-distance plot):

1. Для каждой точки вычислить расстояние до её $k$-го ближайшего соседа (обычно $k = $ `min_samples`).
2. Отсортировать эти расстояния по убыванию и построить график.
3. Найти **«колено»** (knee) графика — точку резкого перегиба. Расстояние в этой точке является хорошим кандидатом для `eps`.

- До колена: точки плотных областей — их расстояние до $k$-го соседа невелико.
- После колена: точки разреженных областей или выбросы — их расстояние резко возрастает.

```python
# Метод k-расстояний для подбора eps на масштабированном Iris
k_neighbors = 5  # обычно равно min_samples

nn = NearestNeighbors(n_neighbors=k_neighbors)
nn.fit(X_iris_scaled)

# Расстояния до k-го ближайшего соседа
distances, _ = nn.kneighbors(X_iris_scaled)
k_distances = distances[:, -1]  # расстояние до k-го соседа

# Сортировка по убыванию
k_distances_sorted = np.sort(k_distances)[::-1]

plt.figure(figsize=(10, 5))
plt.plot(k_distances_sorted, linewidth=2)
plt.axhline(y=0.75, color='red', linestyle='--', alpha=0.8, label='eps=0.75 (колено)')
plt.xlabel('Точки (отсортированные по убыванию расстояния)')
plt.ylabel(f'Расстояние до {k_neighbors}-го соседа')
plt.title(f'Метод k-расстояний (k={k_neighbors}) — подбор eps для Iris')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print(f"Рекомендуемое значение eps (по колену): ~0.5")
print(f"Диапазон k-расстояний: [{k_distances_sorted.min():.3f}, {k_distances_sorted.max():.3f}]")
print(f"Медиана k-расстояний: {np.median(k_distances_sorted):.3f}")
```

## Влияние масштабирования

DBSCAN использует расстояния для определения окрестности, поэтому **масштаб признаков напрямую влияет на результат**. Признаки с большим диапазоном будут «растягивать» пространство и искажать плотностную структуру данных.

Сравним кластеризацию Iris с применением `StandardScaler` и без него при одинаковых параметрах.

```python
# DBSCAN без масштабирования
dbscan_noscale = DBSCAN(eps=0.5, min_samples=5)
dbscan_noscale.fit(X_iris)
labels_noscale = dbscan_noscale.labels_

n_clusters_noscale = len(set(labels_noscale)) - (1 if -1 in labels_noscale else 0)
n_noise_noscale = np.sum(labels_noscale == -1)

# DBSCAN со масштабированием (уже вычислено выше)
n_clusters_scale = n_clusters_iris
n_noise_scale = n_noise_iris

ari_noscale = adjusted_rand_score(y_iris, labels_noscale)
ari_scale = adjusted_rand_score(y_iris, labels_iris)

print("Сравнение DBSCAN с масштабированием и без (eps=0.5, min_samples=5):")
print(f"\n  Без масштабирования: кластеров={n_clusters_noscale}, шума={n_noise_noscale}, ARI={ari_noscale:.4f}")
print(f"  С масштабированием:  кластеров={n_clusters_scale}, шума={n_noise_scale}, ARI={ari_scale:.4f}")
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

datasets = [
    (labels_noscale, X_iris, f'Без масштабирования\nкластеров: {n_clusters_noscale}, шума: {n_noise_noscale}, ARI={ari_noscale:.3f}'),
    (labels_iris, X_iris, f'С масштабированием (StandardScaler)\nкластеров: {n_clusters_scale}, шума: {n_noise_scale}, ARI={ari_scale:.3f}')
]

for ax, (labels, X_data, title) in zip(axes, datasets):
    unique_labels = sorted(set(labels))
    color_map = plt.cm.tab10(np.linspace(0, 1, max(len([l for l in unique_labels if l >= 0]), 1)))
    cluster_idx = 0

    for label in unique_labels:
        mask = labels == label
        if label == -1:
            ax.scatter(X_data[mask, 2], X_data[mask, 3],
                       c='grey', marker='x', s=80, alpha=0.8, label='Шум', zorder=5)
        else:
            ax.scatter(X_data[mask, 2], X_data[mask, 3],
                       color=color_map[cluster_idx], edgecolor='black', s=60, alpha=0.8,
                       label=f'Кластер {label}')
            cluster_idx += 1

    ax.set_xlabel('Длина лепестка (petal length)')
    ax.set_ylabel('Ширина лепестка (petal width)')
    ax.set_title(title)
    ax.legend()

plt.suptitle('Влияние масштабирования на DBSCAN (Iris, eps=0.5, min_samples=5)', fontsize=13)
plt.tight_layout()
plt.show()
```

## Выводы

- **DBSCAN** — алгоритм кластеризации на основе плотности, не требующий заранее задавать число кластеров.
- В отличие от K-Means, DBSCAN находит кластеры **произвольной формы** (в том числе невыпуклые, как полумесяцы) и явно помечает выбросы меткой $-1$.
- **eps** — ключевой параметр: малое значение приводит к дроблению кластеров и избыточному шуму, большое — к слиянию всех точек в один кластер. Метод k-расстояний помогает выбрать eps по «колену» графика.
- **min_samples** контролирует «строгость» ядровости: высокие значения увеличивают долю шума и выделяют только самые плотные области.
- **Масштабирование признаков обязательно** — при одинаковом eps данные без масштабирования и после него дают принципиально разные результаты (разное число кластеров и качество по ARI).
- На датасете Iris DBSCAN с eps=0.5 и min_samples=5 после масштабирования находит структуру, близкую к реальным классам, хотя вид setosa и пара versicolor/virginica могут объединяться в зависимости от параметров.
- Основное ограничение DBSCAN — **плохая работа с кластерами разной плотности**: единые eps и min_samples не могут одновременно подходить для плотных и разреженных групп.
